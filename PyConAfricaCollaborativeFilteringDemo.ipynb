```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")

movies.head()
```

```python
movies = pd.DataFrame({
    'movieId': [1, 2, 3, 4, 5],
    'title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']
})

# Ratings DataFrame: userId, movieId, rating
ratings = pd.DataFrame({
    'userId': [1, 1, 1, 2, 2, 2, 3, 3, 4, 4, 5, 5],
    'movieId': [1, 2, 3, 2, 3, 4, 1, 3, 2, 5, 4, 5],
    'rating': [4, 5, 3, 2, 4, 3, 5, 2, 3, 4, 1, 5]
})
```

```python
final_dataset = ratings.pivot(index='movieId',columns='userId',values='rating')
final_dataset.head()
```

```python
final_dataset.fillna(0,inplace=True)
final_dataset.head()
```

```python

csr_data = csr_matrix(final_dataset.values)
final_dataset.reset_index(inplace=True)
```

```python
print(csr_data)
```

```python
nn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
nn.fit(csr_data)
```

```python
def get_movie_recommendation(movie_name):
    n_movies_to_reccomend = 2
    movie_list = movies[movies['title'].str.contains(movie_name)]
    if len(movie_list):
        movie_idx= movie_list.iloc[0]['movieId']
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        distances , indices = nn.kneighbors(csr_data[movie_idx],n_neighbors=n_movies_to_reccomend+1)
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        recommend_frame = []
        for val in rec_movie_indices:
            movie_idx = final_dataset.iloc[val[0]]['movieId']
            idx = movies[movies['movieId'] == movie_idx].index
            recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0],'Distance':val[1]})
        df = pd.DataFrame(recommend_frame,index=range(1,n_movies_to_reccomend+1))
        return df
    else:
        return "No movies found. Please check your input"
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances

# Sample data creation for movies and ratings (if you have your own dataset, you can load it instead)
# Movies DataFrame: movieId, title
movies = pd.DataFrame({
    'movieId': [1, 2, 3, 4, 5],
    'title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']
})

# Ratings DataFrame: userId, movieId, rating
ratings = pd.DataFrame({
    'userId': [1, 1, 1, 2, 2, 2, 3, 3, 4, 4, 5, 5],
    'movieId': [1, 2, 3, 2, 3, 4, 1, 3, 2, 5, 4, 5],
    'rating': [4, 5, 3, 2, 4, 3, 5, 2, 3, 4, 1, 5]
})

# 1. Create a movie-user matrix
movie_user_matrix = ratings.pivot(index='movieId', columns='userId', values='rating').fillna(0)

# 2. Compute similarity between movies using cosine similarity
movie_similarity = cosine_similarity(movie_user_matrix)

# 3. Create a DataFrame to store similarity scores
movie_similarity_df = pd.DataFrame(movie_similarity, index=movies['title'], columns=movies['title'])

# 4. Define a function to get top N recommendations based on a movie title
def get_movie_recommendations(movie_title, top_n=5):
    """
    Given a movie title, this function returns the top N most similar movie recommendations
    based on collaborative filtering.

    :param movie_title: Title of the movie to find similar movies for.
    :param top_n: Number of recommendations to return.
    :return: List of (movie_title, similarity_score) for the top similar movies.
    """
    if movie_title not in movie_similarity_df.index:
        return f"Movie '{movie_title}' not found in the dataset."

    # Get the similarity scores for the given movie
    similar_movies = movie_similarity_df[movie_title].sort_values(ascending=False)

    # Exclude the movie itself from the recommendations
    similar_movies = similar_movies.drop(movie_title)

    # Get the top N similar movies
    top_recommendations = similar_movies.head(top_n)

    # Return the list of movie titles and their similarity scores
    return list(top_recommendations.items())

# Example usage:
recommendations = get_movie_recommendations('Movie A', top_n=3)
print("Recommendations for 'Movie A':", recommendations)
```
